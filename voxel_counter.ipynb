```python
import os
import numpy as np
import seaborn as sns
import matplotlib
import h5py
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd

data_folder = Path("S:/2020-10-27CH4993_PDFCT_metallate/A178/map")
firstfile = 0
lastfile = 27
q = firstfile-1
v =2 #PDFCT experiment number
thr = 0.6 #threshold for counting voxels
scan_num = '%d' % v
name = 'Gofr_test_BiVO4_'+scan_num+'_'
#new lines for array
cols = ['bivo', 'bi_nano', 'libi_nano', 'li3bi_nano']
lst = []

#chan = 107
#upper_lim = 6 #coke vmax=6, a-b vmax=0.18

while q<lastfile:
    q=q+1
    formatted_q ='%03d' % q 
    # set nuber of digits with the '%03d' where 03 = 3 digits, i.e. 1 becomes 001.
    
    drut = 'A178-XRD-CT_{}_{}_0_sino_tmean20_rec_transform.h5'
    grut = drut.format(scan_num, formatted_q)
    filename = data_folder / grut
    f = h5py.File(filename, 'r')
    data = f['gr']
    #tsst=ds['chan', 10:90, 10:90]
    #create tomograms for the selected G(r) channels
    bivo=data[18, 10:90, 10:90]
    bi_nano=data[15,10:90,10:90]
    libi_nano=data[17,10:90,10:90]
    li3bi_nano=data[24,10:90,10:90]
    bi_bulk=data[32,10:90,10:90]
    libi_bulk=data[57,10:90,10:90]
    li3bi_bulk=data[62,10:90,10:90]
   #count the values in the matrix above a threshold  "thr" 
    over1 = np.sum(bivo > thr) 
    over2 = np.sum(bi_nano > thr) 
    over3 = np.sum(libi_nano > thr)
    over4 = np.sum(li3bi_nano > thr)
#append these to a numpy list
    lst.append([over1, over2, over3, over4]) 
#after the loop, turn the whole list into a Pandas dataframe
df1 = pd.DataFrame(lst, columns=cols)
#normalise the columns in the dataframe
normalized_df1=(df1-df1.min())/(df1.max()-df1.min())
#save the normalised and  raw dataframes as csv files
drit = 'A178-XRD-CT_{}_voxel_counts_norm.csv'
srit = 'A178-XRD-CT_{}_voxel_counts.csv'
grit = drit.format(scan_num)
brit = srit.format(scan_num)
normalized_df1.to_csv(grit)
df1.to_csv(brit)
```
